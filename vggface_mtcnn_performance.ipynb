```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
```

```python
models = [
  "VGG-Face", 
  "Facenet", 
  "Facenet512", 
  "OpenFace", 
  "DeepFace", 
  "DeepID", 
  "ArcFace", 
  "Dlib", 
  "SFace",
]
```

```python
backends = [
  'opencv', 
  'ssd', 
  'dlib', 
  'mtcnn', 
  'retinaface', 
  'mediapipe'
]
```

```python
metrics = ["cosine", "euclidean", "euclidean_l2"]
```

```python
model_name = models[0]
backend_name = backends[3]
metric_name = metrics[2]
```

```python
person_no = 1
filter_no = 9
```

```python
original_img = "Dataset/Face Dataset/{}/{}.jpg".format(person_no, person_no)
filtered_img = "Dataset/Face Dataset/{}/Filtered/{}_{}.jpg".format(person_no, person_no, filter_no)
```

```python
img1 = cv2.imread(original_img)
plt.imshow(img1[:, :, ::-1])
```

```python
img2 = cv2.imread(filtered_img)
plt.imshow(img2[:, :, ::-1])
```

```python
res = DeepFace.verify(img1_path = original_img, 
                      img2_path = filtered_img, 
                      model_name = model_name, 
                      detector_backend = backend_name, 
                      distance_metric = metric_name,
                      enforce_detection=False)
```

```python
res
```

```python
db_path = "Dataset/Face Dataset/Base Images"
tp,total = 0,0
incorrect = []
missed = []
for person_no in range(1, 103):
    for filter_no in range(1, 11):
        filtered_img = "Dataset/Face Dataset/{}/Filtered/{}_{}.jpg".format(person_no, person_no, filter_no)
        df = DeepFace.find(img_path = filtered_img, 
                           db_path = db_path, 
                           model_name=model_name, 
                           detector_backend=backend_name, 
                           distance_metric = metric_name,
                           enforce_detection=False)
        if df.shape[0] != 0:
            strng = df.iloc[0].identity
            i = len(strng)-1
            while i >= 0 and strng[i] != '/':
                i = i-1
            predicted_label = int(strng[i+1:len(strng)-4])
            if predicted_label == person_no:
                tp += 1
            else:
                incorrect.append({'Image': filtered_img, 'Prediction': predicted_label})
        else:
            missed.append({'Image': filtered_img, 'Prediction': -1})
        total += 1
```

```python
print("Percentage of correct predictions for model {} and detector {} is {}%".format(model_name, backend_name, (tp/total)*100))
```
